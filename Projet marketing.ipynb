```python
RECEIPTS_INFO = RECEIPTS_INFO.drop("REMISE",axis=1)
```

```python
RECEIPTS_INFO['REMISE_VALEUR'] = RECEIPTS_INFO['REMISE_VALEUR'].apply(lambda x: 100 if x > 100 else x)
```

```python
RECEIPTS_INFO['PLV'] = RECEIPTS_INFO['MODELE'].apply(lambda x: 1 if x == 'PLV' else 0)
```

```python
RECEIPTS_INFO['Gift'] = RECEIPTS_INFO.apply(lambda row: 1 if (row['MODELE'] in ['FAVO', 'FAVORI']) and (row['PRIX_AP_REMISE'] == 0) else 0, axis=1)
```

```python
RECEIPTS_INFO['Entry-level'] = RECEIPTS_INFO['MODELE'].apply(lambda x: 1 if x == 'ACCESS' else 0)
```

```python
RECEIPTS_INFO_OK = RECEIPTS_INFO.copy()
```

```python
RECEIPTS_INFO_OK["DATE_ACHAT"] = pd.to_datetime(RECEIPTS_INFO_OK["DATE_ACHAT"],format='"%d/%m/%Y"')
```

```python
RECEIPTS_INFO_OK["PRICE"] = RECEIPTS_INFO_OK["PRIX_AP_REMISE"]*RECEIPTS_INFO_OK["QUANTITE"]
```
