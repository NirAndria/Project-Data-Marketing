```python
VISIT_VALUE = (
    RECEIPTS_INFO_RFM
    .groupby(["DATE_ACHAT","CODE_BOUTIQUE","ID_INDIVIDU",  "NUM_TICKET"])
    .aggregate(NB_PRODUCTS=pd.NamedAgg(column="PRIX_AP_REMISE",aggfunc="sum")
    ,VISIT_VALUE=pd.NamedAgg(column="QUANTITE",aggfunc="sum"))
    .reset_index()
)
```

```python
VISIT_VALUE.columns
```

```python
VISIT_VALUE.columns = [
    'DATE_ACHAT', 'CODE_BOUTIQUE', 'ID_INDIVIDU',
    'NUM_TICKET', 'VISIT_VALUE', 'NB_PRODUCTS'
]
```

```python
VISIT_VALUE["AVG_PRICE"] = VISIT_VALUE["VISIT_VALUE"]/VISIT_VALUE["NB_PRODUCTS"]
```

```python
VISIT_VALUE.head(100)
```

```python
VISIT_VALUE[(VISIT_VALUE.CODE_BOUTIQUE == 130) & (VISIT_VALUE.ID_INDIVIDU==156593)]
```

```python
agg_dict = {"ID_INDIVIDU":"count",
            "VISIT_VALUE":["mean","sum"],
            "NB_PRODUCTS":"mean"
            
}

FREQUENCY = (
    VISIT_VALUE
    .groupby("ID_INDIVIDU")
    .aggregate(agg_dict, drop_index = True)
    .reset_index()
)
```

```python
FREQUENCY.columns = ['ID_INDIVIDU','NB_VISITS','AVG_VISIT_VALUE','CUMUL_VALUE','AVG_NB_PRODUCTS_VISIT']
```

```python
FREQUENCY
```

```python
RECENCY = (
    VISIT_VALUE
    .groupby(["ID_INDIVIDU"])
    .aggregate(Last=("DATE_ACHAT", "max"))
    .reset_index()
)
```

```python
RECENCY.head()
```

```python
RECENCY["RECENCY"] = RECEIPTS_INFO_RFM.DATE_ACHAT.max()-RECENCY["Last"]
```

```python
RECENCY.head()
```

```python
DIVERS_STORE = (
    RECEIPTS_INFO_RFM
    .groupby('ID_INDIVIDU')
    .aggregate(NB_STORES=pd.NamedAgg(column="CODE_BOUTIQUE",aggfunc="count"))
    .reset_index()
)

DIVERS_STORE.columns = ['ID_INDIVIDU', 'NB_STORES']
```

```python
DIVERS_FAM = (
    RECEIPTS_INFO_RFM
    .groupby('ID_INDIVIDU')
    .aggregate(NB_FAMILIES=pd.NamedAgg(column="ID_FOYER",aggfunc="count"))
    .reset_index()
)
DIVERS_FAM.columns = ['ID_INDIVIDU', 'NB_FAMILIES']
```

```python
DIVERS_LINE = (
    RECEIPTS_INFO_RFM
    .groupby('ID_INDIVIDU')
    .aggregate(NB_LINES=pd.NamedAgg(column="Ligne",aggfunc="count"))
    .reset_index()
)
DIVERS_LINE.columns = ['ID_INDIVIDU', 'NB_LINES']
DIVERS_FAM.head()
```

```python
NB_GIFTS = (
    RECEIPTS_INFO_RFM
    .groupby('ID_INDIVIDU')
    .aggregate(NB_GIFTS=pd.NamedAgg(column="Gift",aggfunc="sum"))
    .reset_index()
)

NB_GIFTS.columns = ['ID_INDIVIDU', 'NB_GIFTS']
NB_GIFTS.head()
```
